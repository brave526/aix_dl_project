```python
dataset = pd.concat([apt_avg, kospi_kosdaq2, nasdaq2, rate2],axis=1)
```

```python
dataset
```

```python
dataset.columns = ['날짜', '평균매매가격', '코스피지수', '코스닥지수', '나스닥종가', '국고채3년평균', '국고채5년평균', '국고채10년평균', '회사채3년평균', '기준금리']
```

```python
x = dataset[['코스피지수', '코스닥지수', '나스닥종가', '국고채3년평균', '국고채5년평균', '국고채10년평균', '회사채3년평균', '기준금리']]
y = dataset[['평균매매가격']]

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(x_train, y_train)

mlr_result = mlr.fit(x_train, y_train)
```
